```python
# 1.For each movie, compute the number of cast members
```

```python
#Definiamo una funzione che carichi in una variabile Y  i file json della variabile X e che ritorni la lunghezza di Y.
def CastMembers(X):
    Y = json.loads(X)
    return(len(Y))
#Applichiamo la funzione alla colonna 'cast' del database Credits, ed inseriamo i risultati nella nuova colonna 'Cast Members'.
credits['CastMembers'] = credits['cast'].apply(CastMembers)
credits[['movie_id','title','CastMembers']].head()
```

```python
# 4.Extract the domain of each homepage
```

```python
ConHomepage = movies[movies.homepage.notnull()]
ConHomepage = ConHomepage.reset_index()
def domaine(hp):
    domains = []
    homepage = hp.split(" ")
    for i in range(len(homepage)):
        domains.append(homepage[i].split('/')[2].split('.')[-1])
    return str(domains)[1:-1].replace("\'", "")

ConHomepage["domains"]=ConHomepage["homepage"].apply(domaine)
ConHomepage[["title","homepage","domains"]]
```

```python
# 5.Extract a set of normalized tables. That is, each entry of a normalized table must contain exactly one value
# (not a list or a dictionary). There is no need to use SQL for this point.
```

```python
def normalizza (movie_id, colonna, dataset):
    temp = dataset.set_index(movie_id)[colonna].apply(pd.read_json)
    def setta_chiave(indice):
        temp[indice][movie_id] = indice
        return temp
    dataset[movie_id].apply(setta_chiave)
    return pd.concat(temp.values.tolist())
```

```python
movies_n = movies.drop(['genres', 'keywords', 'production_companies', 'production_countries', 'spoken_languages'], axis = 1)
credits_n = credits.drop(['cast', 'crew'], axis = 1)
genres = normalizza('movie_id','genres', movies)
keywords = normalizza('movie_id', 'keywords', movies)
production_companies = normalizza('movie_id','production_companies', movies)
production_countries = normalizza('movie_id', 'production_countries', movies).rename(columns = {'iso_3166_1':'country_id'})
spoken_languages = normalizza('movie_id', 'spoken_languages', movies).rename(columns = {'iso_639_1':'language_id'})
crew = normalizza('movie_id','crew', credits)
cast = normalizza('movie_id','cast', credits)
```

```python
# 7.For each movie, compute the number of crew members
```

```python
crew2 = crew[["id", "movie_id"]].drop_duplicates()
crew_count = pd.DataFrame(crew2.groupby("movie_id").size()).reset_index()
crew_count.columns = ["movie_id", "crew_count"]
crew_count = pd.merge(crew_count, movies, left_on="movie_id", right_on="movie_id")
crew_count[["title", "crew_count"]].head()
```

```python
# 8.For each movie, compute the number of directors
```

```python
crew3 = crew[crew.job == "Director"]
directors_count = crew3.groupby("movie_id").size().reset_index()
directors_count.columns = ["movie_id", "directors_count"]
directors_count = pd.merge(directors_count, movies, left_on="movie_id", right_on="movie_id")
directors_count[["title", "directors_count"]].sort_values("directors_count", ascending=0).head()
```

```python
# 10.For each company and each decade, compute the overall revenue
```

```python
production_companies_name = production_companies[['id','name']].drop_duplicates()
```

```python
movies["year"] = movies["release_date"].str.extract("(.*)-.*-.*",expand=True)

def isdecade(year):
    return str(year)[:3] + '0'
movies['decade'] = movies['year'].apply(isdecade)
companies_id = production_companies.merge(right=movies[['movie_id','revenue','decade']], on='movie_id')
companies_name = production_companies_name.merge(right=companies_id[['id','revenue','decade']], on='id')
tot_revenue=companies_name.groupby(by =['id','name','decade'], as_index=False).sum()
tot_revenue
```

```python
# 11.For each decade, compute the company with maximum revenue.
```

```python
best_revenue10= tot_revenue.groupby(by =['decade'])[["revenue"]].apply(max, axis = 0).reset_index()
best_revenue = best_revenue10["revenue"].values
decade = best_revenue10["decade"].values

for i in range(len(best_revenue10)):
    print(tot_revenue[tot_revenue["decade"]==decade[i]][tot_revenue["revenue"]==best_revenue[i]],"\n")
```
